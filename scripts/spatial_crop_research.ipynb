```python
from imagebind.data import load_and_transform_video_data

video_path = "../datasets/AVIClips/actionclipautoautotrain00018.avi"
video_input, before_crop = load_and_transform_video_data(
    video_paths=[video_path],
    device='cpu',
)
```

```python
before_crop[0].shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_image(image_data: np.ndarray):
    if image_data.shape[0]:
        image_data = image_data.transpose(1, 2, 0)
    fig = plt.figure(frameon=False)
    ax = plt.Axes(fig, [0.0, 0.0, 1.0, 1.0])
    ax.set_axis_off()
    fig.add_axes(ax)
    ax.imshow(
        image_data,
    )
    
draw_image(before_crop[0][:, 1, :, :].numpy())
```

```python
idx = 6
draw_image(
    video_input[0, idx, :, 0, ...].numpy()
)

# video_input.shape
```
